```python
query = "SELECT * FROM customer_churn LIMIT 5"

result = pd.read_sql(query, connection)

result
```

```python
query = """
SELECT customerID,
       gender,
       MonthlyCharges
FROM customer_churn;
"""

pd.read_sql(query, connection)
```

```python
query = """
SELECT customerID,
       MonthlyCharges
FROM customer_churn;
"""

result = pd.read_sql(query, connection)

result
```

```python
query = """
SELECT *
FROM customer_churn
WHERE Churn='Yes';
"""

pd.read_sql(query, connection)
```

```python
query = """
SELECT COUNT(*) AS Total_Customers
FROM customer_churn;
"""

pd.read_sql(query, connection)
```

```python
query = """
SELECT AVG(MonthlyCharges) AS Average_Monthly_Charge
FROM customer_churn;
"""

pd.read_sql(query, connection)
```

```python
query = """
SELECT Contract,
       COUNT(*) AS Customers
FROM customer_churn
GROUP BY Contract;
"""

pd.read_sql(query, connection)
```
